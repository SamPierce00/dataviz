```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn. datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns

iris = load_iris()
X = iris.data
y = iris.target
scaler = StandardScaler()
Scaled_iris=scaler.fit_transform(X)    
pca_iris = PCA()
x_new = pca_iris.fit_transform(X)
iris_loadings = pca_iris.components_
```

```python
iris_df = pd.DataFrame (x_new[:, :2], columns= ['first', 'second'])
iris_df['class']= y
sns.relplot(data= iris_df, x= 'first', y= 'second', hue= 'class' ,height= 10, aspect= 1.5, palette='Paired' , s= 60)
labels= list(iris.feature_names)
for i in range (4):
    plt.arrow(0, 0, iris_loadings[0,i], iris_loadings[1,i],capstyle='projecting', 
               edgecolor='#ba2121ff', facecolor='#ba2121ff', linewidth=2, head_width=.1)
    plt.text(iris_loadings[0,i]* 1.6, iris_loadings[1,i] * 1.6, labels[i], color = 'purple', ha = 'center', va = 'center')
plt.axis('equal')
plt.xlabel("First Component")
plt.ylabel("Second Component")
plt.title("Biplot for Iris Data");
```
