```python
import pandas as pd
```

```python
df= pd.read_csv ('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
plt.figure (figsize= (15,10))
plt.hist(df.age)
plt.xlabel ('Age');
```

```python
fig, ax = plt.subplots (4,3, figsize= (30,20) )
ax[0,0].hist(df.age, bins= 20)
ax[0,1].hist(df.age, bins= 5)
ax[0,2].hist(df.age, bins = 15)
ax[1,0].hist(df.age, bins = 15, alpha= 0.5)
ax[1,1].hist(df.age, bins = 15, density = True)
ax[1,2].hist(df.age, bins = 15, cumulative = True)
ax[2,0].hist(df.age, bins = 15, orientation= 'horizontal')
ax[2,1].hist(df[df['sex']=='male'].age, bins = 15, alpha = 0.7, edgecolor= 'gray')
ax[2,1].hist(df[df['sex']=='female'].age, bins = 15, alpha = 0.9, edgecolor= 'gray')
ax[2,2].hist([df[df['sex']=='male'].age, df[df['sex']=='female'].age], bins = 15, edgecolor= 'gray' , stacked= True)
ax[3,0].hist(df[df['sex']=='female'].age, bins = 15, orientation= 'horizontal')
ax[3,0].hist(-df[df['sex']=='male'].age, bins = 15, orientation= 'horizontal')
ax[3,1].hist(-df[df['sex']=='male'].age, bins = 15)
ax[3,1].set_xticks(ticks= [-60, -40, -20, 0, 20, 40, 60], labels= [60, 40, 20, 0, 20, 40, 60])
ax[3,1].hist(df[df['sex']=='female'].age, bins = 15)
;
```

## Class Activity: Try to make the age pyramid?

```python
import seaborn as sns
sns.kdeplot(df.age ,color='green',shade=True, bw_adjust= 0.6)
plt.xlabel ('Age');
```

```python
sns.kdeplot(df[df['sex']=='male'].age ,color='green',shade=True, bw_adjust= 0.6)
sns.kdeplot(df[df['sex']=='female'].age ,color='blue',shade=True, bw_adjust= 0.6)
plt.xlabel ('Age');
```

```python
sns.kdeplot(df[df['sex']=='male'].age ,color='green',shade=True, bw_adjust= 0.6, common_norm=False)
sns.kdeplot(df.age ,color='gray',shade=True, bw_adjust= 0.6, common_norm= False);
```
